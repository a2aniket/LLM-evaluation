```python

from langchain_community.document_loaders import PyPDFLoader
from langchain_community.document_loaders import PyPDFDirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceBgeEmbeddings

EMBEDDING_MODEL_NAME: str = "BAAI/bge-small-en-v1.5"
EMBEDDING_PATH: str = "misc/embeddings/"
MODEL_KWARGS: dict = {'device': 'cpu'}
ENCODE_KWARGS: dict = {'normalize_embeddings': True}

file="./General principles of fractures treatment 1.pdf"
loader = PyPDFLoader(file)
documents = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
final_documents = text_splitter.split_documents(documents)

huggingface_embeddings = HuggingFaceBgeEmbeddings(
        model_name=EMBEDDING_MODEL_NAME,
        model_kwargs=MODEL_KWARGS,
        encode_kwargs=ENCODE_KWARGS
    )
retrive = FAISS.from_documents(final_documents, huggingface_embeddings)

qa = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retrive.as_retriever(),
    return_source_documents=True,
)
```

```python
qa({"query": "principles of fractures treatment"})
```

```python
def model(input_df):
    answer = []
    for index, row in input_df.iterrows():
        answer.append(qa(row["questions"]))

    return answer
```

```python
import pandas as pd

eval_df = pd.DataFrame(
    {
        "questions": [
            "What is a fracture?",
            "How to classify fractures?",
            "What are the clinical features of fractures?",
            "What are the principles of fracture treatment?",
            "How to manage open fractures?",
            "What are the complications of fracture treatment?",
            "How does fracture healing work?",
            "What are the types of internal fixation?",
            "When is external fixation indicated?",
            "What are the different cast complications?"
        ],
    }
)
```

# Faithfulness

faithfulness ensures the output is accurate and grounded in the source

```python
from mlflow.metrics.genai import EvaluationExample, faithfulness

# Create a good and bad example for faithfulness in the context of this problem

faithfulness_examples = [
    EvaluationExample(
        input="What are the main principles of fracture treatment?",
        output="The main principles of fracture treatment are anatomical reduction, stable fixation, preservation of blood supply, and early mobilization. The goal is to obtain union of the fracture in the most anatomical position for maximal functional return.",
        score=5,
        justification="The output accurately reflects the information provided in the context about principles of treatment without adding unsupported claims.",
        grading_context={
            "context": "Principles of treatment • Anatomical reduction • Stable internal fixation • Preservation of blood supply • Early mobilization [...] The goal of fracture treatment is to obtain union of the fracture in the most anatomical position compatible with maximal functional return of the extremity."
        }
    ),
    EvaluationExample(
        input="What are the main principles of fracture treatment?",
        output="Fracture treatment involves surgery and physical therapy. Patients should take calcium supplements and do weight-bearing exercises after 2 weeks of rest.",
        score=1,
        justification="The output makes claims about calcium supplements and specific timeline recommendations that are not present in the provided context, demonstrating poor faithfulness to the source material.",
        grading_context={
            "context": "Principles of treatment • Anatomical reduction • Stable internal fixation • Preservation of blood supply • Early mobilization [...] The goal of fracture treatment is to obtain union of the fracture in the most anatomical position compatible with maximal functional return of the extremity."
        }
    )
]

faithfulness_metric = faithfulness(model="openai:/WM16KModel", examples=faithfulness_examples)
```

#Relevance
1. refers to how well a model's output aligns with the context, query, or intended
2. relevance focuses on the output's usefulness, appropriateness, and adherence to the task's intent. It is a critical metric in tasks like information retrieval, dialogue generation, summarization, and question answering.



```python
from mlflow.metrics.genai import EvaluationExample, relevance

relevance_examples = [
    EvaluationExample(
        input="What are the main principles of fracture treatment?",
        output="The main principles of fracture treatment are anatomical reduction, stable fixation, preservation of blood supply, and early mobilization. The goal is to obtain union of the fracture in the most anatomical position for maximal functional return.",
        score=5,
        justification="The output is highly relevant to the input question, directly addressing the principles of fracture treatment without introducing unrelated details.",
        grading_context={
            "context": "Principles of treatment • Anatomical reduction • Stable internal fixation • Preservation of blood supply • Early mobilization [...] The goal of fracture treatment is to obtain union of the fracture in the most anatomical position compatible with maximal functional return of the extremity."
        }
    ),
    EvaluationExample(
        input="What are the main principles of fracture treatment?",
        output="Fractures are common injuries often treated by orthopedic specialists. Modern medicine has advanced significantly in fracture care.",
        score=2,
        justification="The output provides generic information about fractures but fails to address the specific principles of treatment mentioned in the input question, making it only marginally relevant.",
        grading_context={
            "context": "Principles of treatment • Anatomical reduction • Stable internal fixation • Preservation of blood supply • Early mobilization [...] The goal of fracture treatment is to obtain union of the fracture in the most anatomical position compatible with maximal functional return of the extremity."
        }
    ),
    EvaluationExample(
        input="What are the main principles of fracture treatment?",
        output="The diet of a patient with a fracture should include protein-rich foods and vitamin D supplements.",
        score=1,
        justification="The output is irrelevant to the input question as it discusses diet rather than the principles of fracture treatment, which is the focus of the query.",
        grading_context={
            "context": "Principles of treatment • Anatomical reduction • Stable internal fixation • Preservation of blood supply • Early mobilization [...] The goal of fracture treatment is to obtain union of the fracture in the most anatomical position compatible with maximal functional return of the extremity."
        }
    )
]

relevance_metric = relevance(model="openai:/WM16KModel", examples=relevance_examples)

```

```python
results = mlflow.evaluate(
    model,
    eval_df,
    model_type="question-answering",
    evaluators="default",
    predictions="result",
    extra_metrics=[faithfulness_metric, relevance_metric, mlflow.metrics.latency()],
    evaluator_config={
        "col_mapping": {
            "inputs": "questions",
            "context": "source_documents",
        }
    },
)
print(results.metrics)
```

#Retriever Metrics

```python
import pandas as pd

# Create a sample dataset
data = {
    "inputs": [
        "What is the capital of France?",
        "Explain Newton's first law",
        "Benefits of regular exercise",
        "Python vs Java for beginners",
        "Latest trends in AI"
    ],
    "ground_truth": [
        ["doc1", "doc2"],  # Relevant documents for the first query
        ["doc3"],          # Relevant documents for the second query
        ["doc4", "doc5"],  # Relevant documents for the third query
        ["doc6"],          # Relevant documents for the fourth query
        ["doc7", "doc8"]   # Relevant documents for the fifth query
    ],
    "retrieved_docs": [
        ["doc1", "doc9", "doc2"],  # Retrieved documents for the first query
        ["doc10", "doc3", "doc11"],# Retrieved documents for the second query
        ["doc5", "doc12", "doc13"],# Retrieved documents for the third query
        ["doc14", "doc6", "doc15"],# Retrieved documents for the fourth query
        ["doc7", "doc16", "doc8"]  # Retrieved documents for the fifth query
    ]
}

# Convert the dictionary to a Pandas DataFrame
df = pd.DataFrame(data)

# Display the dataset
print(df)
```

```python
def retriever_function(query_df):
  retrive_1 = retrive.as_retriever(
        search_kwargs={"k": 1 }
    )
  qa = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retrive_1,
    return_source_documents=True,
  )

  all_retrieved_doc_ids = []
  for index, row in query_df.iterrows():
      query_text = row["inputs"]
      retrieved_documents = qa({"query": query_text})["source_documents"]
      retrieved_doc_ids = [doc.metadata["source"] for doc in retrieved_documents]
      all_retrieved_doc_ids.append(retrieved_doc_ids)

  return all_retrieved_doc_ids  # Return a list of lists
```

```python
result=mlflow.evaluate(
    model=retriever_function,
    data=df,
    targets="ground_truth",
    model_type="retriever",
    evaluators="default",
    evaluator_config={"retriever_k": 1}
)
print(result.metrics)
```

```python
result.tables["eval_results_table"]
```

```python
# with a static dataset and using `extra_metrics`
result=mlflow.evaluate(
    data=df,
    predictions="retrieved_docs", # Changed from "predictions_param" to "retrieved_docs"
    targets="ground_truth",      # Changed from "targets_param" to "ground_truth"
    model_type="retriever",
    extra_metrics=[
        mlflow.metrics.precision_at_k(5),
        mlflow.metrics.precision_at_k(6),
        mlflow.metrics.recall_at_k(5),
        mlflow.metrics.ndcg_at_k(5)
    ]
)
```
